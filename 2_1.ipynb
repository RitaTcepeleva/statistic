```python
import scipy.stats as sps
import numpy as np
import random
import ipywidgets as widgets
import matplotlib.pyplot as plt
%matplotlib inline

A = 4
B = 8

m = A+B
sigma = np.abs(A-B)
N = 60

first = []
for i in range(N):
    first.append(random.uniform(0,1))

print(first)
```

```python
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

```python
norm_first = []

for i in range(N):
    norm_first.append(sps.norm.ppf(first[i], loc=m, scale = sigma))

print(norm_first)
```

Еще одна выборка:

```python
norm_second = sps.norm(m, sigma).rvs(size=N)
print(norm_second)
```

Вычисляем выборочные коэффициенты корреляции (Пирсон):

```python
pearsons_cor_coef23 = sps.pearsonr(norm_first, norm_second)
print("Коэф коррел 23: ",pearsons_cor_coef23[0])

pearsons_cor_coef13 = sps.pearsonr(first, norm_second)
print("Коэф коррел 13: ",pearsons_cor_coef13[0])

pearsons_cor_coef12 = sps.pearsonr(first, norm_first)
print("Коэф коррел 12: ",pearsons_cor_coef12[0])
```

Вычислите несмещённые оценки ковариации для трёх пар выборок (1,2), (2, 3), (1, 2).
Вычислите средние квадратические отклонения для выборок и проверьте по формуле парного
коэффициента корреляции полученные ранее с помощью функции Excel значения
коэффициентов корреляции:

```python
mean1 = np.mean(first)
mean2 = np.mean(norm_first)
mean3 = np.mean(norm_second)

cov12 = 0
for i in range(N):
    cov12 += (first[i] - mean1)*(norm_first[i] - mean2)
cov12 = cov12 / (N-1)
print("Ковариация 12: ",cov12)

cov13 = 0
for i in range(N):
    cov13 += (first[i] - mean1)*(norm_second[i] - mean3)
cov13 = cov13 / (N-1)
print("Ковариация 13: ",cov13)

cov23 = 0
for i in range(N):
    cov23 += (norm_first[i] - mean2)*(norm_second[i] - mean3)
cov23 = cov23 / (N-1)
print("Ковариация 23: ",cov23)

std1 = np.std(first, ddof = 1)
std2 = np.std(norm_first, ddof = 1)
std3 = np.std(norm_second, ddof = 1)
print("Ср кв откл 1: ",std1)
print("Ср кв откл 2: ",std2)
print("Ср кв откл 3: ",std3)

numerator = 0
sum1 = 0
sum2 = 0
for i in range(N):
    numerator += (first[i] - mean1)*(norm_first[i] - mean2)
    sum1 += (first[i] - mean1) ** 2
    sum2 += (norm_first[i] - mean2) ** 2

denominator = np.sqrt(sum1 * sum2)

r12 = numerator / denominator
print("Коэф коррел 12: ",r12)
print("Значение, посчитанное выше по формуле: ", pearsons_cor_coef12[0])

numerator = 0
sum1 = 0
sum2 = 0
for i in range(N):
    numerator += (first[i] - mean1)*(norm_second[i] - mean3)
    sum1 += (first[i] - mean1) ** 2
    sum2 += (norm_second[i] - mean3) ** 2

denominator = np.sqrt(sum1 * sum2)

r13 = numerator / denominator
print("Коэф коррел 13: ",r13)
print("Значение, посчитанное выше по формуле: ", pearsons_cor_coef13[0])

numerator = 0
sum1 = 0
sum2 = 0
for i in range(N):
    numerator += (norm_first[i] - mean2)*(norm_second[i] - mean3)
    sum1 += (norm_first[i] - mean2) ** 2
    sum2 += (norm_second[i] - mean3) ** 2

denominator = np.sqrt(sum1 * sum2)

r23 = numerator / denominator
print("Коэф коррел 23: ",r23)
print("Значение, посчитанное выше по формуле: ", pearsons_cor_coef23[0])
```
